```python
import numpy as np
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm

import os

# print(model.wv['0'])
# print(model.wv.vocab['0'])

from os import listdir
from os.path import isfile, join
onlyfiles = [f for f in os.listdir(dir_name) if isfile(join(dir_name, f)) and f.endswith('.vect')]

for database_name in onlyfiles: 
    model = Word2Vec.load(os.path.join(dir_name, database_name))
    data = [(model.wv.vocab[x].count, np.linalg.norm(model.wv[x])) for x in model.wv.vocab]

    x = np.array([x[0] for x in data]).reshape((-1, 1))
    y = np.array([x[1] for x in data])
    
    print(f'database: {database_name}')
    print (sorted([x for x in data][:10], key=lambda x:x[0], reverse=True))

    mod = sm.OLS(y,x)
    fii = mod.fit()
    p_values = fii.summary2().tables[1]

  
    
    print(p_values)
```
